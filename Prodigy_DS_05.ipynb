Analyze traffic accident data and their causes to identify patterns and Visualize.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('road_accidents.csv')
```

```python
df
```

### Data Understanding
sno	- Serial number of the state/UT (Integer)

stateut	- Name of the state/UT (String)

region - Region of the state/UT (String)

alcintake2014 - Number of accidents due to alcohol intake in 2014 (Integer)

overspeed2014 - Number of accidents due to overspeeding in 2014 (Integer)

overtaking2014 - Number of accidents due to overtaking in 2014 (Integer)

lanejumping2014 - Number of accidents due to lane jumping in 2014 (Integer)

asleep2014 - Number of accidents due to sleeping at the wheel in 2014 (Integer)

othercause2014 - Number of accidents due to miscellaneous causes in 2014 (Integer)

alcintake2016 - Number of accidents due to alcohol intake in 2016 (Integer)

overspeed2016 - Number of accidents due to overspeeding in 2016 (Integer)

lanejumping2016 - Number of accidents due to lane jumping in 2016 (Integer)

overtaking2016 - Number of accidents due to overtaking in 2016 (Integer)

asleep2016 - Number of accidents due to sleeping at the wheel in 2016 (Integer)

othercause2016 - Number of accidents due to miscellaneous causes in 2016 (Integer)

### Looking for Outliers and imputing them

```python
df.columns
```

```python
cols=df[['regionid', 'alcintake2014', 'overspeed2014',
       'overtaking2014', 'lanejumping2014', 'wrongside2014', 'signalavoid2014',
       'asleep2014', 'othercause2014', 'alcintake2016', 'overspeed2016',
       'signalavoid2016', 'wrongside2016', 'lanejumping2016', 'overtaking2016',
       'asleep2016', 'othercause2016']]

plt.figure(figsize=(20,50))
plotno = 1

for column in cols:
    if plotno<=14:   
        plt.subplot(7,2,plotno) 
        sns.boxplot(cols[column]) 
        plt.xlabel(column,fontsize=20) 
    plotno+=1 
plt.tight_layout()
plt.show() 
```

### Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler() 
df[['regionid', 'alcintake2014', 'overspeed2014',
       'overtaking2014', 'lanejumping2014', 'wrongside2014', 'signalavoid2014',
       'asleep2014', 'othercause2014', 'alcintake2016', 'overspeed2016',
       'signalavoid2016', 'wrongside2016', 'lanejumping2016', 'overtaking2016',
       'asleep2016', 'othercause2016']]=scalar.fit_transform(df[['regionid', 'alcintake2014', 'overspeed2014',
       'overtaking2014', 'lanejumping2014', 'wrongside2014', 'signalavoid2014',
       'asleep2014', 'othercause2014', 'alcintake2016', 'overspeed2016',
       'signalavoid2016', 'wrongside2016', 'lanejumping2016', 'overtaking2016',
       'asleep2016', 'othercause2016']])
```

### Visualizations

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='alcintake2014', y='alcintake2016', data=df)
plt.title('Alcohol Intake vs Accident Occurrence (2014 vs 2016)')
plt.xlabel('Alcohol Intake 2014')
plt.ylabel('Alcohol Intake 2016')
plt.show()
```

```python
correlation_matrix = df[['overspeed2014', 'overtaking2014', 'lanejumping2014', 'wrongside2014',
                          'signalavoid2014', 'asleep2014', 'othercause2014']].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Contributing Factors (2014)')
plt.show()
```

```python
correlation_matrix = df[['alcintake2016', 'overspeed2016',
       'signalavoid2016', 'wrongside2016', 'lanejumping2016', 'overtaking2016',
       'asleep2016', 'othercause2016']].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Contributing Factors (2016)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['alcintake2014'], bins=20, kde=True, color='blue', label='2014')
sns.histplot(df['alcintake2016'], bins=20, kde=True, color='red', label='2016')
plt.title('Distribution of Alcohol Intake in 2014 and 2016')
plt.xlabel('Alcohol Intake')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
sns.pairplot(data=df[['overspeed2014', 'overtaking2014', 'lanejumping2014', 'wrongside2014',
                      'signalavoid2014', 'asleep2014', 'othercause2014']])
plt.suptitle('Pairplot of Contributing Factors (2014)', y=1.02)
plt.show()
```

```python
sns.pairplot(data=df[['alcintake2016', 'overspeed2016',
       'signalavoid2016', 'wrongside2016', 'lanejumping2016', 'overtaking2016',
       'asleep2016', 'othercause2016']])
plt.suptitle('Pairplot of Contributing Factors (2016)', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='stateut', y='alcintake2014', data=df, ci=None)  # ci=None disables error bars
plt.title('Average Alcohol Intake in 2014 by State/UT')
plt.xlabel('State/UT')
plt.ylabel('Average Alcohol Intake 2014')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='stateut', y='alcintake2016', data=df, ci=None)  # ci=None disables error bars
plt.title('Average Alcohol Intake in 2016 by State/UT')
plt.xlabel('State/UT')
plt.ylabel('Average Alcohol Intake 2016')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['alcintake2014', 'alcintake2016', 'overspeed2014', 'overspeed2016',
                      'overtaking2014', 'overtaking2016', 'lanejumping2014', 'lanejumping2016',
                      'wrongside2014', 'wrongside2016']], orient='h')
plt.title('Boxplot Matrix of Numerical Variables')
plt.xlabel('Value')
plt.ylabel('Variable')
plt.show()
```
